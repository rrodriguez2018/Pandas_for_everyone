```python
import pandas as pd
billboard = pd.read_csv('../data/billboard.csv')
```

```python
billboard_long = billboard.melt(
    id_vars=["year","artist","track","time","date.entered"],
    var_name="week",
    value_name="rating",
)
billboard_long
```

```python
# suppose we subset data based on a particular track:
billboard_long.loc[billboard_long.track == 'Loser']
```

```python
billboard_songs = billboard_long[
    ["year","artist","track","time"]
]
billboard_songs.shape
```

```python
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
# by adding 1 to the id column, we start with 1 instead of zero
billboard_songs['id'] = billboard_songs.index + 1
billboard_songs
```

```python
# with new ID column created, we can match a song to its weekly ranking
billboard_ratings = billboard_long.merge(
    billboard_songs, on=["year","artist","track","time"]
)
billboard_ratings.shape
```

```python
billboard_ratings
```

```python
# Finally, we subset the columns to the ones we want in our ratings dataframe
billboard_ratings = billboard_ratings[
    ["id","date.entered","week","rating"]
]
billboard_ratings
```
