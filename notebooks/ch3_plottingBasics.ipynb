```python
import seaborn as sns
```

```python
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
import matplotlib.pyplot as plt
dataset1 = anscombe[anscombe['dataset']== 'I']
dataset1
```

```python
plt.plot(dataset1['x'], dataset1['y'])
plt.show()
```

```python
plt.plot(dataset1['x'], dataset1['y'],'*')
plt.show()
```

```python
dataset2 = anscombe[anscombe['dataset'] == 'II']
dataset3 = anscombe[anscombe['dataset'] == 'III']
dataset4 = anscombe[anscombe['dataset'] == 'IV']
dataset3
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

# add a plot to each of the figures created above
axes1.plot(dataset1['x'],dataset1['y'],'o')
axes2.plot(dataset2['x'],dataset2['y'],'o')
axes3.plot(dataset3['x'],dataset3['y'],'o')
axes4.plot(dataset4['x'],dataset4['y'],'o')

# add a label to each sub plot & add xlabel & y-label
axes1.set_title("data set_1")
axes1.set_xlabel("time 1")
axes1.set_ylabel("money 1")
axes2.set_title("data set_2")
axes2.set_xlabel("time 2")
axes2.set_ylabel("money 2")
axes3.set_title("data set_3")
axes3.set_xlabel("time 3")
axes3.set_ylabel("money 3")
axes4.set_title("data set_4")
axes4.set_xlabel("time 4")
axes4.set_ylabel("money 4")

fig.suptitle("Anscombe Data")
fig.set_tight_layout(True)

plt.show()
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.hist(data=tips, x='total_bill', bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
plt.show()
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(data=tips, x='total_bill',y='tip')
axes1.set_title('Scatter of Total Bill Vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
plt.show()
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot(
    x=[
        tips.loc[tips["sex"] == "Female", "tip"],
        tips.loc[tips["sex"] == "Male", "tip"],
    ],
    labels=["Female", "Male"],
)
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Gender')
plt.show()

```

```python
colors = {           # we map the colors to each value of sex:
    "Female": "#f1a340",
    "Male": "#998ec3",
}
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(data=tips, x='total_bill',y='tip',s=tips["size"] ** 2 * 10,c=tips["sex"].map(colors),alpha=0.5)
# we need to set the size of the dots based on party size, so we multiply the values by 10

axes1.set_title('Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.suptitle("Total Bill Vs Tips")
plt.show()
```

```python
fig,ax = plt.subplots()  # this is the Pandas plotting method (simpler)
tips['total_bill'].plot.hist(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots()
tips[['total_bill', 'tip']].plot.hist(alpha=0.5, bins=20, ax=ax)
plt.show()
```
